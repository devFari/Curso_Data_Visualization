```python
#importações

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import dados

url_csv = 'https://raw.githubusercontent.com/devFari/Curso_Data_Visualization/refs/heads/main/dados/imigrantes_canada.csv'

df = pd.read_csv(url_csv)
df
```

```python
df.info()
```

```python
#mudar o index para a coluna País
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
```

```python
#criando um dicionario
brasil_dic = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dic)
```

```python
dados_brasil
```

```python
#construindo grafico
plt.figure(figsize=(10, 5)) #modifica o tamanho do grafico (largura, haltura)
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) #(dados do eixo X, dados do eixo Y)
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Qtd Imigrantes')

#definindo os marcadores do eixo X
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
#definindo os marcadores do eixo Y
plt.yticks([500 , 1000, 1500, 2000, 2500, 3000])
plt.show()
```

```python
#criando figura para fazer mais altreações no grafico
fig, ax = plt.subplots(figsize=(10, 5)) #definindo o tamanho da figura
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes']) #definindo os dados do grafico, eixo X e Y
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013') #definido p titulo, OBS. quando usando '\n' significa que estamos indo para a proxima linha
ax.set_xlabel('ano') #definindo o rotulo do eixo x
ax.set_ylabel('Qtd de imigrantes') #definindo o rotulo do eixo Y
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #definindo o intervalo do eixo X
plt.show()
```

```python
#para gerar mais de um grafico na mesma figura'AXS'
fig, axs = plt.subplots(1, 2, figsize=(15,5)) #1argumento quantas linhas , 2 argumento quantos graficos

#primeiro grafico
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o canadá\n1980 a 2013')
axs[0].set_xlabel('ano')
axs[0].set_ylabel('Qtd de Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid() #adiciona uma grade

#segundo grafico
axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da Imigração do Brasil para o Canada\n1980 a 2013')
axs[1].set_xlabel('brasil')
axs[1].set_ylabel('Nº de Imigrantes')
axs[1].grid() #adiciona uma grade

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,7))
fig.subplots_adjust(hspace=0.5, wspace=0.3) #Ajusta o espaçamento entre os plots
fig.suptitle('Imigração dos quatro maiores países da America do Sul para o Canadá de 1980 a 2013')

#grafico 01
axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

#grafico 02
axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

#grafico 03
axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

#grafico 04
axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

#loop que vai percorrer todo o codigo
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #espaçamento dos anos

# definindo rotulos genericos
for ax in axs.flat:
  ax.set(xlabel='Ano', ylabel='Qtd de Imigrantes')

#definindo limite maximo e minimo
ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

```python
#criando figura para fazer mais altreações no grafico 02
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, color='g') #lw autera a espeçura da linha | 'marker' adiciona marcadores | 'color' modifica a cor do grafico
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=18, loc='left') #fontsize altera o tamanho da fonte | loc autera a localização do titulo
ax.set_xlabel('Ano',fontsize=14)
ax.set_ylabel('Qtd de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12) #auterando o tamanho do ticck do eixo X
ax.yaxis.set_tick_params(labelsize=12) #auterando o tamanho do tick do eixo Y
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
#plt.grid(linestyle='--') # 'lineStyle' modifica a forma da linha do grid
plt.show()
```
